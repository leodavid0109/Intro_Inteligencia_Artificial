```python
def ordenar_pila(pila):
    """
    Ordena una pila utilizando pilas auxiliares e imprime los pasos realizados.
    :param pila: Lista que representa la pila a ordenar (elemento superior es el último de la lista).
    :return: Pila ordenada (de menor a mayor, con el menor en el fondo).
    """
    pilas_auxiliares = [[]]  # Lista que puede contener múltiples pilas auxiliares
    pasos = 0

    print("Estado inicial:")
    print("Pila principal:", pila)
    print("Pilas auxiliares:", pilas_auxiliares)

    while pila:
        # Sacar el elemento del tope de la pila original
        elemento_actual = pila.pop()

        # Seleccionar la primera pila auxiliar o crear una nueva si es necesario
        pila_actual = pilas_auxiliares[-1]
        while pila_actual and pila_actual[-1] > elemento_actual:
            pila.append(pila_actual.pop())
            pasos += 1
            print(f"Paso {pasos}:")
            print("Pila principal:", pila)
            print("Pilas auxiliares:", pilas_auxiliares)

        # Si la pila actual no puede aceptar el elemento_actual, crear una nueva pila auxiliar
        if pila_actual and pila_actual[-1] <= elemento_actual:
            pila_actual.append(elemento_actual)
        else:
            pilas_auxiliares.append([elemento_actual])

        pasos += 1
        print(f"Paso {pasos}:")
        print("Pila principal:", pila)
        print("Pilas auxiliares:", pilas_auxiliares)

    # Transferir los elementos de las pilas auxiliares a la pila principal en orden
    while pilas_auxiliares:
        pila_actual = pilas_auxiliares.pop()
        while pila_actual:
            pila.append(pila_actual.pop())
            pasos += 1
            print(f"Paso {pasos}:")
            print("Pila principal:", pila)
            print("Pilas auxiliares:", pilas_auxiliares)

    print("Pila ordenada:", pila)
    print(f"Total de pasos: {pasos}")

    return pila

# Ejemplo de uso
pila = ['B', 'D', 'C', 'A']
print("Pila original:", pila)
pila_ordenada = ordenar_pila(pila)
print("Pila ordenada:", pila_ordenada)
```
